```python
x=heart_disease[["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal"]]
print(x.to_string())
```

```python
y=heart_disease[["target"]]
print(y.to_string())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)#random_state=to maintain stable accuracy at every time
```

**DecisionTree**

```python
from sklearn.metrics import accuracy_score
from sklearn import tree
dt=tree.DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

**SVM**

```python
from sklearn.metrics import accuracy_score
from sklearn import svm
clf=svm.SVC()
clf.fit(x_train,y_train)
y_pred= clf.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

**Randomforest**

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
rf.fit(x_train,y_train)
y_pred= rf.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
clf=svm.SVC()
clf.fit(x_train,y_train)
y_pred= clf.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)

#confusion matrix
cm=confusion_matrix(y_pred,y_test)

#plot
plt.figure(figsize=(6,5))
sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='RdPu',
            cbar=True,
            linewidths=2,
            linecolor='black')
#labels
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
